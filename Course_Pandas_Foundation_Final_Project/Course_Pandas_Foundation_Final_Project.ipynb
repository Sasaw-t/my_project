```python
# TODO 04 - your friend ask for all order data in `California` and `Texas` in 2017 (look at Order Date), send him csv file
ct2017_data = df.query('(State == "California" | State == "Texas") & `Order Date`.dt.year == 2017')
ct2017_data.to_csv('ct2017_data.csv')
ct2017_data
```

```python
# TODO 05 - how much total sales, average sales, and standard deviation of sales your company make in 2017
df_2017 = df[df['Order Date'].dt.year == 2017]
df_2017.groupby(df['Order Date'].dt.to_period('M'))['Sales'].agg(['sum', 'mean', 'std'])
```

```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df['Order Date'].dt.year == 2018]
df_2018.groupby('Segment')['Profit'].sum().sort_values(ascending=False).head(1)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
start_date = pd.to_datetime('2019-04-15')
end_date = pd.to_datetime('2019-12-31')

filtered_df = df[(df['Order Date'] >= start_date) & (df['Order Date'] <= end_date)]
filtered_df.groupby('State')['Sales'].sum().sort_values(ascending=True).head()
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25%
df_2019 = df[df['Order Date'].dt.year == 2019]
total_sales_2019 = df_2019['Sales'].sum()

wc_df_2019 = df_2019.query('Region == "West" | Region == "Central"')
wc_sales_2019 = wc_df_2019['Sales'].sum()

wc_sales_percentage_2019 = (wc_sales_2019 / total_sales_2019) * 100
print(f"proportion of total sales in West + Central in 2019 = {wc_sales_percentage_2019:.2f}%")
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020
filtered_date_df = df.query('`Order Date`.dt.year == 2019 | `Order Date`.dt.year == 2020')

top_10_num_orders = filtered_date_df.groupby('Product Name').size().sort_values(ascending=False).head(10)

print("Top 10 Products by Number of Orders in 2019-2020")
print(top_10_num_orders)
print("\n" + "-"*50 + "\n")

top_10_total_sales = filtered_date_df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

print("Top 10 Products by Total Sales in 2019-2020")
print(top_10_total_sales)
```

```python
# TODO 10 - plot at least 2 plots, any plot you think interesting :)
# 1.Plot profits in each region

df.groupby('Region')['Profit'].sum().sort_values(ascending=False).plot(kind='bar');
```

```python
# 2.Plot monthly profit
df_month_profit = df.groupby(df['Order Date'].dt.to_period('M'))['Profit'].sum()
df_month_profit.plot(kind='line', figsize=(12, 4));
```

```python
# TODO Bonus - use np.where() to create new column in dataframe to help you answer your own questions
#            - Use np.where() to check if a state has a positive or negative average profit

import numpy as np

df_mean_state_profit = df.groupby('State')['Profit'].mean()
df_state_profit = df_mean_state_profit.to_frame()

df_state_profit['Profit Status'] = np.where(df_state_profit['Profit'] >= 0, 'Positive', 'Negative')
df_state_profit
```
